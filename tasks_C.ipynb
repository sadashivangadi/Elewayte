```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Housing.csv")

# Convert rupees → lakhs
df['price_lakhs'] = df['price'] / 100000

ranges = {
    "0–25 Lakhs": (0, 25),
    "26–50 Lakhs": (26, 50),
    "51–75 Lakhs": (51, 75),
    "76–100 Lakhs": (76, 100),
    ">100 Lakhs": (101, 10000)
}

labels = []
counts = []

for label, (low, high) in ranges.items():
    count = df[(df['price_lakhs'] >= low) & (df['price_lakhs'] <= high)].shape[0]
    labels.append(label)
    counts.append(count)

print("House Price Ranges and their Counts")
for l, c in zip(labels, counts):
    print(f"{l}: {c}")

# Plot
plt.figure(figsize=(8,5))
plt.plot(labels, counts, marker='o')
plt.xlabel("Price Range (Lakhs)")
plt.ylabel("Number of Houses")
plt.title("House Count by Price Range")
plt.grid(True)
plt.show()
```
